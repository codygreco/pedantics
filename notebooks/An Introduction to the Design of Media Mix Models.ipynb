```python
import pygraphviz as pgv
from IPython.display import Image

G = pgv.AGraph(directed=True, rankdir="LR")

G.add_node("Node A")
G.add_node("Node B")
G.add_node("Node C")

G.add_edge("Node A", "Node B", label="Directed Edge AB")

G.add_edge("Node B", "Node C", label="Undirected Edge BC", dir="none")
G.add_edge("Node A", "Node A", label="Cycle AA", dir="none")
G.add_edge("Node C", "Node A", label="Cycle CA", dir="none")


graph_file_path = "graph_horizontal.png"
G.draw(graph_file_path, format="png", prog="dot")

Image(graph_file_path)
```

```python
G = pgv.AGraph(directed=True)

nodes = ["Pricing", "Promotion", "Product", "Revenue"]

for node in nodes:
    G.add_node(node)

edges = [("Pricing", "Sales"),
         ("Promotion", "Sales"),
         ("Product", "Sales")]

for edge in edges:
    G.add_edge(edge[0], edge[1])

graph_file_path = "causal_graph.png"
G.draw(graph_file_path, format="png", prog="dot")

Image(graph_file_path)
```

```python

G = pgv.AGraph(directed=True)

nodes = [
    "Paid Search",
    "Organic Search",
    "Awareness Advertising",
    "Sales",
    "Pricing",
    "Promotion",
    "Unobserved Variables",
    "Seasonality",
    "Market Trends",
    "Competitor Actions",
]

for node in nodes:
    G.add_node(node)

edges = [
    ("Awareness Advertising", "Paid Search"),
    ("Awareness Advertising", "Organic Search"),
    ("Organic Search", "Paid Search"),
    ("Organic Search", "Sales"),
    ("Paid Search", "Sales"),
    ("Pricing", "Sales"),
    ("Promotion", "Sales"),
    ("Unobserved Variables", "Awareness Advertising"),
    ("Unobserved Variables", "Paid Search"),
    ("Unobserved Variables", "Organic Search"),
    ("Unobserved Variables", "Sales"),
    ("Seasonality", "Awareness Advertising"),
    ("Seasonality", "Paid Search"),
    ("Seasonality", "Organic Search"),
    ("Seasonality", "Sales"),
    ("Market Trends", "Awareness Advertising"),
    ("Market Trends", "Paid Search"),
    ("Market Trends", "Sales"),
    ("Competitor Actions", "Awareness Advertising"),
    ("Competitor Actions", "Paid Search"),
    ("Competitor Actions", "Sales"),
]

for edge in edges:
    G.add_edge(edge[0], edge[1])

graph_file_path = "complex_causal_graph.png"
G.draw(graph_file_path, format="png", prog="dot")

Image(graph_file_path)
```

```python
G = pgv.AGraph(directed=True)

nodes = [
    "Economic Factors",
    "Consumer Demand",
    "Non-Search Contributors",
    "Search Queries",
    "Auction",
    "Budget",
    "Organic Search",
    "Search Ad",
    "Sales",
]

for node in nodes:
    G.add_node(node)

edges = [
    ("Economic Factors", "Consumer Demand"),
    ("Consumer Demand", "Non-Search Contributors"),
    ("Non-Search Contributors", "Sales"),
    ("Consumer Demand", "Sales"),
    ("Consumer Demand", "Search Queries"),
    ("Non-Search Contributors", "Search Queries"),
    ("Search Queries", "Auction"),
    ("Auction", "Search Ad"),
    ("Search Ad", "Sales"),
    ("Budget", "Non-Search Contributors"),
    ("Budget", "Search Ad"),
    ("Search Queries", "Organic Search"),
    ("Organic Search", "Sales"),
]

for edge in edges:
    G.add_edge(edge[0], edge[1])

graph_file_path = "search_causal_graph.png"
G.draw(graph_file_path, format="png", prog="dot")

Image(graph_file_path)
```
