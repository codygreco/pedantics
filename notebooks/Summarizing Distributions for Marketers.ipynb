```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from numpy.random import Generator, PCG64
import seaborn as sns
```

```python
lower, upper = 0, 100000
mu, sigma = 10000, 1000
lower = (lower-mu)/sigma
upper = (upper - mu) / sigma
X = stats.truncnorm.rvs(a=lower, b=upper, loc=mu, scale=sigma, size=335, random_state=1352)
```

```python
import numpy as np
rng = np.random.RandomState(42)
random_outliers = rng.randint(10000, 16000, size=30)
```

```python
new_X = np.concatenate([X, random_outliers])
```

```python
fig, ax = plt.subplots()
sns.histplot(new_X, ax=ax)
ax.set_xlabel("Revenue")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of Revenue on Marketing Campaign")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.histplot(new_X, ax=ax, stat='probability')
ax.set_xlabel("Revenue")
ax.set_ylabel("Probability")
ax.set_title("Distribution of Revenue on Marketing Campaign")
plt.show()
```

```python
campaign0 = stats.skewnorm.rvs(a=20, loc=5000, scale=1000, size=305, random_state=1352)
campaign1 = stats.norm.rvs(loc=5000, scale=10000, size=100, random_state=123123)
campaign1[campaign1 <0] = 0
```

```python
fig, axes = plt.subplots(figsize=(12, 4), ncols=2)
sns.histplot(campaign0, ax=axes[0])
sns.histplot(campaign1, ax=axes[1])

axes[0].set_xlabel("Revenue")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Distribution of Revenue on Marketing Campaign One\nStandard Deviation: {0}".format(campaign0.std().round(2)))

axes[1].set_xlabel("Revenue")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Distribution of Revenue on Marketing Campaign two\nStandard Deviation: {0}".format(campaign1.std().round(2)))
plt.show()
```
