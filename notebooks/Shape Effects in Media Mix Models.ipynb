```python
marketing_spend = np.linspace(0, 10, 100)

def sigmoid_curve(x):
    return 1 / (1 + np.exp(-x))

def concave_curve(x):
    return np.sqrt(x)

def convex_curve(x):
    return x**2

def linear_curve(x):
    return x

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(marketing_spend, sigmoid_curve(marketing_spend), label='S-shape (Sigmoid)', color="salmon")
plt.title('Marketing Response: S-shape (Sigmoid)')

plt.subplot(2, 2, 2)
plt.plot(marketing_spend, concave_curve(marketing_spend), label='Concave', color="salmon")
plt.title('Marketing Response: Concave')

plt.subplot(2, 2, 3)
plt.plot(marketing_spend, convex_curve(marketing_spend), label='Convex', color="salmon")
plt.title('Marketing Response: Convex')

plt.subplot(2, 2, 4)
plt.plot(marketing_spend, linear_curve(marketing_spend), label='Linear', color="salmon")
plt.title('Marketing Response: Linear')
plt.suptitle("Proposed Shape Effects in Marketing", size=14)
plt.tight_layout()
plt.show()
```

```python
days = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")
daily_data = np.random.normal(0, 1, len(days)) + 0.2 * np.sin(np.arange(len(days)))
daily_data[100] += 10  # Adding an outlier
daily_data = pd.DataFrame(daily_data, index=days)

weekly_data = daily_data.resample("W").mean()
monthly_data = daily_data.resample("M").mean()
quarterly_data = daily_data.resample("Q").mean()

plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(days, daily_data, label='Daily Data', color='salmon')
plt.title('Daily Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(weekly_data.index, weekly_data, label='Weekly Aggregation', color='salmon')
plt.title('Weekly Aggregated Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(monthly_data.index, monthly_data, label='Monthly Aggregation', color='salmon')
plt.title('Monthly Aggregated Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(quarterly_data.index, quarterly_data, label='Quarterly Aggregation', color='salmon')
plt.title('Quarterly Aggregated Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
num_points = 365
spend_values = np.linspace(0, 1000, num_points)

linear_campaign = spend_values + np.random.normal(0, 50, num_points)
saturating_campaign_1 = 200 * (1 - np.exp(-spend_values * 0.005)) + np.random.normal(0, 20, num_points)
saturating_campaign_2 = 300 * (1 - np.exp(-spend_values * 0.007)) + np.random.normal(0, 30, num_points)

overall_effect = linear_campaign + saturating_campaign_1 + saturating_campaign_2

data = pd.DataFrame({
    "spend": spend_values,
    "linear_campaign": linear_campaign,
    "saturating_campaign_1": saturating_campaign_1,
    "saturating_campaign_2": saturating_campaign_2,
    "total": overall_effect
})

plt.figure(figsize=(12, 8))

plt.plot(data["spend"], data["linear_campaign"], label='Linear Campaign', color='salmon')
plt.plot(data["spend"], data["saturating_campaign_1"], label='Saturating Campaign 1', color='teal')
plt.plot(data["spend"], data["saturating_campaign_2"], label='Saturating Campaign 2', color='lightgrey')
plt.plot(data["spend"], data["total"], label='Overall Effect', linestyle='--', color='black')

plt.title('The Effects of Aggregating Marketing Data')
plt.xlabel('Spend')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(0, 10, 200)

def hill_function(x, v_max, Km, n):
    return v_max * (x**n) / ((Km**n) + (x**n))

plt.figure(figsize=(10, 6))

plt.plot(x_values, hill_function(x_values, 1, 1, 1), label='v_max=1, Km=1, n=1', color='salmon')

plt.plot(x_values, hill_function(x_values, 1, 2, 3), label='v_max=1, Km=2, n=3', color='teal')

plt.plot(x_values, hill_function(x_values, 2, 1, 1), label='v_max=2, Km=1, n=1', color='lightgrey')

plt.title('Hill Functions with Different Parameter Values')
plt.xlabel('x')
plt.ylabel('Hill Function Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
campaign_data = pd.read_csv("campaign_data.csv")
```

```python
def hill_function(spend, Vmax, Km, n):
    return Vmax * (spend / (Km + spend)) ** n

def calculate_rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))


def linear_regression(x, m, b):
    return m * x + b
    
selected_campaigns_dict = {}
bad_error = 0
good_error =0
# Loop over  campaign_ids
for campaign_id in campaign_data['id'].unique():
    pivot_df = campaign_data[campaign_data['id'] == campaign_id].pivot_table(
        index="date", values="metric_value", columns="metric_name"
    )
    pivot_df = pivot_df[pivot_df.SPEND > 0]
    pivot_df.fillna(0, inplace=True)
    # Sort data by spend
    pivot_df = pivot_df.sort_values(by="SPEND")
    scaler = MinMaxScaler()
    scaled_data = pd.DataFrame(
        scaler.fit_transform(pivot_df[["SPEND", "CHANNEL_REPORTED_REVENUE"]]),
        columns=["SPEND", "CHANNEL_REPORTED_REVENUE"],
    )
    scaled_data.sort_values(by='SPEND', inplace=True)
    split_index = int(len(scaled_data) * 0.5)
    train_data, test_data = (
        scaled_data.iloc[:split_index],
        scaled_data.iloc[split_index:],
    )

    # Split data into features and target
    spend_train, revenue_train = (
        train_data["SPEND"].values,
        train_data["CHANNEL_REPORTED_REVENUE"].values,
    )
    spend_test, revenue_test = (
        test_data["SPEND"].values,
        test_data["CHANNEL_REPORTED_REVENUE"].values,
    )

    params, covariance = curve_fit(
        hill_function, spend_train, revenue_train, maxfev=10000
    )
    Vmax_fit, Km_fit, n_fit = params
    hill_train_pred = hill_function(spend_test, Vmax_fit, Km_fit, n_fit)

    linear_model = LinearRegression()
    linear_model.fit(spend_train.reshape(-1, 1), revenue_train)
    m, b = linear_model.coef_[0], linear_model.intercept_
    linear_train_pred = linear_regression(spend_test, m, b)

    hill_train_error = calculate_rmse(hill_train_pred, revenue_test)
    linear_train_error = calculate_rmse(linear_train_pred, revenue_test)
    
    if linear_train_error < hill_train_error:
        bad_error += 1
        x = np.linspace(0, 1)
        reg = linear_regression(x, m, b)
        hill = hill_function(x, Vmax_fit, Km_fit, n_fit)
        spend_data = scaled_data.SPEND.values
        rev_data = scaled_data.CHANNEL_REPORTED_REVENUE.values
        selected_campaigns_dict[campaign_id] = {
            "reg_res": reg,
            "hill_res": hill,
            "spend_data": spend_data,
            "rev_data": rev_data,
            "hill_error": hill_train_error,
            "lin_error": linear_train_error
        }
    else:
        good_error += 1
```

```python
print(good_error)
print(bad_error)
print(bad_error + good_error)
print((bad_error)/ (good_error + bad_error))
```

```python
error_dfs = []
for campaign in selected_campaigns_dict.keys():
    error_dfs.append(
        pd.DataFrame({
            "lin_error": selected_campaigns_dict[campaign]['lin_error'],
            "hill_error": selected_campaigns_dict[campaign]['hill_error']
        }, index=[campaign])
    )
error_df = pd.concat(error_dfs)
```

```python
error_df['diff'] = error_df['hill_error'] - error_df['lin_error']
error_df = error_df[error_df['diff'] <= 1]
```

```python
plot_campaigns = list(error_df.index)
```

```python
campaign_keys = random.sample(plot_campaigns, 4)
```

```python
campaign_keys = random.sample(plot_campaigns, 4)

def plot_campaign(ax, spend_data, rev_data, reg_res, hill_res, lin_error, hill_error, campaign_id):
    ax.scatter(spend_data, rev_data, color='black', alpha=0.5, label='Data')
    ax.plot(np.linspace(0, 1), reg_res, color='salmon', label=f'Linear Regression\nRMSE: {lin_error:.2f}')
    ax.plot(np.linspace(0, 1), hill_res, color='teal', label=f'Hill Function\nRMSE: {hill_error:.2f}')
    ax.set_xlabel('Spend')
    ax.set_ylabel('Revenue')
    ax.legend()

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, campaign_id in enumerate(campaign_keys):
    row, col = divmod(i, 2)
    plot_campaign(axs[row, col],
                  selected_campaigns_dict[campaign_id]['spend_data'],
                  selected_campaigns_dict[campaign_id]['rev_data'],
                  selected_campaigns_dict[campaign_id]['reg_res'],
                  selected_campaigns_dict[campaign_id]['hill_res'],
                  selected_campaigns_dict[campaign_id]['lin_error'],
                  selected_campaigns_dict[campaign_id]['hill_error'],
                  campaign_id)

plt.suptitle("Error of Hill Functions vs Linear Regression")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 100, 100)
y = 10 + 0.4*x - 0.003*(x**2) + np.random.normal(scale=5, size=100)

data = pd.DataFrame({'spend': x, 'revenue': y})

knots = [30, 60, 90]

transformed_x = dmatrix("bs(spend, knots=knots, degree=3, include_intercept=False)",
                        {"spend": data['spend'], "knots": knots}, return_type='dataframe')

model = sm.GLM(data['revenue'], transformed_x).fit()

data['pred'] = model.predict(transformed_x)

plt.figure(figsize=(10, 6))
plt.scatter(data['spend'], data['revenue'], label='Revenue', color='black')
plt.plot(data['spend'], data['pred'], color='salmon', label='Predicted Revenue - Spline Fit')
plt.title('Spline Fit to Marketing Spend Data')
plt.xlabel('Spend')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
